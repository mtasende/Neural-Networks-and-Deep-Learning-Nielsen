# ¿Por qué las redes profundas (deep networks) son difíciles de entrenar?

Al entrenar redes profundas con SGD, se ve que algunas de las capas aprenden bien mientras que otras no. Esto es debido a lo que se conoce como el "Vanishing Gradient Problem".

```python
import mnist_loader
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
```

```python
import network2
net = network2.Network([784, 30, 10])
```

```python
net.SGD(training_data, 30, 10, 0.1, lmbda=5.0,
       evaluation_data=validation_data,
       monitor_evaluation_accuracy=True)
```

Ahora empiezo a agregar capas ocultas...

```python
net = network2.Network([784, 30, 30, 10])
net.SGD(training_data, 30, 10, 0.1, lmbda=5.0,
       evaluation_data=validation_data,
       monitor_evaluation_accuracy=True)
```

```python
net = network2.Network([784, 30, 30, 30, 10])
net.SGD(training_data, 30, 10, 0.1, lmbda=5.0,
       evaluation_data=validation_data,
       monitor_evaluation_accuracy=True)
```

```python
net = network2.Network([784, 30, 30, 30, 30, 10])
net.SGD(training_data, 30, 10, 0.1, lmbda=5.0,
       evaluation_data=validation_data,
       monitor_evaluation_accuracy=True)
```

Se ve que agregar más capas no mejora la exactitud.

Algunas formas de evitar el "Vanishing Gradient problem" conducen al "Exploding Gradient problem", porque en el fondo el problema es que el gradiente es "inestable" en las DNNs.

Por otro lado no es el único problema con las DNNs. Las activaciones sigmoides parecen no ser buenas porque en la capa final tienden a saturar en 0, y eso limita el aprendizaje.
La elección de parámetros iniciales y "momento" cambian mucho el comportamiento en el caso de "momentum-based" gradient descent. Y puede haber otros. Es un campo de investigación por el momento. Sin embargo hay técnicas que sí han dado resultados. Se ven en el capítulo 6.
